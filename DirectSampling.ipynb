```python
import numpy as np
import matplotlib.pyplot as plt
import random

```

```python
# This function accepts the size of the square side 
# and simulates throwing a stone and seeing if the stone lies inside the circle 
def simulate_trials(square_side):
    # Initialize the number of hits to 0
    n_hits = 0

    # Generate two random numbers between 0 and square_side
    x = np.random.uniform(0, square_side)
    y = np.random.uniform(0, square_side)
    
    # Check if the coordinates are inside the circle
    if x**2 + y**2 <= square_side**2:
        # If the point is inside the circle, increment the number of hits
        n_hits += 1

    # Return the number of hits
    return n_hits
```

```python
# This code estimates pi/4 and that is because 
# the ratio of Area of a circle to the area of a square is pi/4
def estimate_pi(square_side, num_runs):
    errors = []
    num_trials_list = []

    # Run the simulation for the given number of runs
    for _ in range(num_runs):
        # Initialize a variable to store the total number of hits
        total_hits = 0

        # Generate a random number of trials between 10000 and 1e5
        num_trials = random.randint(10000, int(1e5))

        # Run the simulation for the given number of trials
        for _ in range(num_trials):
            total_hits += simulate_trials(square_side)

        # Calculate the estimate of pi/4
        estimate_of_piby4 = total_hits / num_trials

        # Calculate the error
        error = abs(estimate_of_piby4 - np.pi/4)

        # Append the error and the number of trials to the respective lists
        errors.append(error)
        num_trials_list.append(num_trials)

    # Plot the errors
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.plot(errors)
    plt.xlabel('Run')
    plt.ylabel('Error')
    plt.title('Error in estimate of pi/4 over multiple runs')

    plt.subplot(1, 2, 2)
    plt.scatter(num_trials_list, errors)
    plt.xlabel('Number of Trials')
    plt.ylabel('Error')
    plt.title('Error vs Number of Trials')

    plt.tight_layout()
    plt.show()

# Call the estimate_pi function with square_side = 10, and num_runs = 100
estimate_pi(10, 100)
```
