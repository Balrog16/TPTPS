```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Consider a 1-D particle moving up and down, rather hit the floor and bounce back up
# The particle is in a box of length L = 1 and the particle is at position x = 0.5
# The particle has constant velocity v = -0.1 in the y direction
# The particle hit the floor at y = 0 at t= 5 and bounce back up without any loss of energy
t = np.arange(0, 10, 0.1)
v = -0.1
y_initial = 0.5
y = y_initial + v * t
y = np.where(y < 0, -y, y)
plt.plot
plt.plot(t, y, 'r')
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.title('Particle moving up and down')
plt.show()
```
